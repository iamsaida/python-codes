```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('/content/weatherAUS.csv')
df
```

```python
x=df['Date'][:10].tolist()
y=df['MaxTemp'][:10].tolist()
xa =[x[i:i + 10] for i in range(0, len(x))]
ya = [y[i:i + 10] for i in range(0, len(y))]
```

```python
fig, axs = plt.subplots(2, 3,figsize=(10,10))
axs[0, 0].plot(xa[0], ya[0])
axs[0, 0].set_title('Plot 1')
axs[0, 1].plot(xa[1], ya[1])
axs[0, 1].set_title('Plot 2')
axs[0, 2].plot(xa[2], ya[2])
axs[0, 2].set_title('Plot 3')
axs[1, 0].plot(xa[3], ya[3])
axs[1, 0].set_title('Plot 4')
axs[1, 1].plot(xa[4], ya[4])
axs[1, 1].set_title('Plot 5')
axs[1, 2].plot(xa[5], ya[5])
axs[1, 2].set_title('Plot 6')
plt.tight_layout()
```

```python
plt.bar( x= df['Date'][:6], height = df['MaxTemp'][:6])
plt.title('Maximum temperature in first six days of december 2008')
plt.xlabel('Date')
plt.ylabel('Maximum Temperature')
plt.tight_layout()
```

```python
import numpy as np
x = df['MaxTemp']
plt.hist(x)
plt.xlabel('Maximum Temperature')
plt.ylabel('No. of places recorded')
plt.title('Histograms (defalut bins)')
plt.show()
```

```python
import numpy as np
x = df['MinTemp']
plt.hist(x, bins=20,color='red')
plt.xlabel('Minimum Temperature')
plt.ylabel('No. of places recorded')
plt.title('Histograms (bins=20)')
plt.show()
```

```python
y = df['MinTemp'][:50]
x = df['Date'][:50]
plt.scatter(x, y, label="Min Temp", color='violet')
plt.ylabel('Minimum Temperature')
plt.xlabel('Date')
plt.title('Minimum temperatures of first week december 2008')
plt.legend()
plt.tight_layout()
```
