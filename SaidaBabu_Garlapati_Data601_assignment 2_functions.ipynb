```python
import csv
```

```python
def second_function(link):
 list1 = []
 with open(link) as file:
  reader=csv.reader(file)
  for row in reader:
    new_list=row[5]
    new=new_list.split(' ')
    for item in new:
      if item not in list1:
        list1.append(item)
    latest=(list1[2:])
    with open('/content/DATA601_SP21/HW-2/data601-file.txt', 'w') as f:
     for item in latest:
        f.write("%s\n" % item)
  print(latest)
  return 'successfully written in data601-file.txt'
```

## Question

By using `csv` module again to read the `test_users.csv` in the `test` folder, find the average age of first 1000 entries for those users who we know their ages.

```python
import csv
```

```python
link='/content/test/test_users.csv'
```

```python
list=[]
latest=[]
new=[]
with open(link) as f:
  reader=csv.reader(f)
  for row in reader: 
    list.append(row[5])
  for ele in list: 
     if ele.strip(): 
        latest.append(ele)
new_list=latest[1:]
sum=0
count=0
for number in new_list:
  if count>=1000:
    break
  new.append(number)
  age=float(number)
  sum+=age
  count+=1
avg=sum//count
print('List of ages of first',count, 'entries: ',new)
print('Sum of ages: ',sum)
print('Average of ages: ',avg)
```
