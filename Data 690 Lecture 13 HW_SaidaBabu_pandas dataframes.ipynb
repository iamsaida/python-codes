### 2. Do the manual pivoting of the data by
  1. Grouping the data by 'Description and District'
  2. Selecting 'Neighborhood'
  3. Aggregating the data using count
  4. `unstack()` the table

```python
neighbor_crime.groupby(['Description','District'])['Neighborhood'].aggregate('count').unstack() 
```

### 3. Using Pandas `pivot_table()` method, produce the same outcome as above (using manual groupby)

```python
neighbor_crime.pivot_table('Neighborhood',columns = 'District', index = 'Description',aggfunc='count')     
```
